# Outliers

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
collist=df.columns.values
ncol=30
nrows=14
plt.figure(figsize=(ncol,3*ncol))
for i in range(0,len(collist)):
    plt.subplot(nrows,ncol,i+1)
    sns.boxplot(df[collist[i]],color='green',orient='h')
    plt.tight_layout()
```

```python
from scipy.stats import zscore
```

Z SCORE

```python
z=np.abs(zscore(df))
z
```

```python
# considering threshold =3
print(np.where(z>3)) # index of outliers
```

```python
z.iloc[27]
```

```python
# droping of outliers
df1=df[(z<3).all(axis=1)]
print("old dataset",df.shape)
print("new dataset",df1.shape)
print("total dropped rows",df.shape[0]-df1.shape[0])
```

```python
x=df1
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
x
```

```python
pd.DataFrame(x).skew()
```

```python
from sklearn.preprocessing import StandardScaler # transforms data such that mean =0 and std dev=1
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
y=df1['Happiness Score']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.metrics import r2_score
lr=LinearRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
print("predicted value:",predlr)
print("Actual value:",y_test)
print('R2 score',r2_score(predlr,y_test))
```
